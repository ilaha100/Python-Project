## Preprocessing (Oyunçu dəyərinin rəqəmə çevrilməsi)

##### Value sutununa baxmaq isteyirem. Sutunun ilk 10 reqemini nece gore bilerem?

```python
df["Value"].head(10)
```

##### Burda Euro isaresini cixaririq, M herfini 1000000-la, K herfini ise 1000-le evezleyirik

```python
df["Value"]=df["Value"].str.replace("€","")
def replace_value(value):
    if 'M' in value:
        return float(value.replace('M',''))*10**6
    elif 'K' in value:
        return float(value.replace('K',''))*10**3
    else:
        return float(value)
```

İndi isə sütunu iki hissəyə böləcək kod yazın və bu iki sütunu bir birinə vuraraq real "Value" sütununu əldə edin (float tipində)

```python
df["Real Value"]=df["Value"].apply(replace_value)
```

##### Menim sutunumda null deyerler ola biler. Men isteyirem ki bunu 0 ile evez edim

```python
df["Real Value"].fillna(0,inplace=True)
```

## Visualization and Analysis

#####  Oyunçuların Pozisiya paylanması (Categorical)

Bir fiqur (10,10) yaradın Bu fiqur daxilində hansı pozisiyadan neçə nəfər uyğun chart vasitəsilə göstərin ("Position" sütunu)

```python
#Oyuncularin pozisiyalari
plt.figure(figsize=(10,7))
plt.ylabel("Oyunçu sayı")
plt.xlabel("Pozisiya")
sns.countplot(x="Position",data=df,palette="flare")
plt.xticks(rotation=90)
plt.show()
```

#### Oyunçuların yaş paylanması (Numeric)

Bir fiqur (10,10) yaradın Bu fiqur daxilində hansı yaş sütununu uyğun chart vasitəsilə göstərin ("Age" sütunu)

```python
plt.figure(figsize=(10,6))
plt.ylabel("Oyunçu sayı")
plt.xlabel("Yaş")
sns.histplot(x="Age",data=df,color="navy")
plt.show()
```

Bir də bunu kde plot vasitəsilə verin

```python
plt.figure(figsize=(10,7))
sns.kdeplot(x="Age",data=df,color="navy")
plt.show()
```

## Dashboard

```python

fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,15))
sns.boxplot(x="International Reputation", y="Potential", data=df, ax=ax[0,0],color="blue")
ax[0][1].pie(x=df["Preferred Foot"].value_counts(),labels=df["Preferred Foot"].value_counts().index,autopct="%.2f",colors=["blue","plum"])
sns.histplot(x="Age", data=df, color="navy", ax=ax[1,0])
sns.lineplot(x="Potential", y="Strength", data=df, ax=ax[1,1])
plt.show()
```
